**NOTE**: The return keyword allows you to actually save the result of the output of a function as a variable. The print() function simply displays the output to you, but doesn't save it for future use

````python
#Printing Fibonacci sequence: [0, 1, 1, 2, 3, 5, 7...]
def getFibonacci (totalNum):
  a = 0
  b = 1
  series = []   
  # Check is n is less
  if totalNum < 0:
      print("Incorrect input")
         
  # Check is n is equal
  elif totalNum == 0:
      series.append(0)
      return series
       
  # Check if n is equal to 1
  elif totalNum == 1:
      series.extend([0, 1])
      return series
  
  else:
      series.extend([0, 1])
      for i in range(2, totalNum):
        c = a + b
        series.append(c)
        a = b
        b = c
      return series
````

````python
ans = getFibonacci(5)
print(type(ans))
print(ans)
````

###lambda Function

A lambda function is a small anonymous function. A lambda function can take any number of arguments, but can only have one expression.

Lambda is designed for coding simple functions, and def handles the larger tasks.

`lambda arguments : expression`

````python
x = lambda a : a + 10
print(x(5))
````

````python
x = lambda a, b, c : a + b * c
print(x(5, 6, 2))
````

###map Function

The map function allows you to "map" a function to an iterable object. That is to say you can quickly call the same function to every item in an iterable, such as a list

````python
def square(num):
    return num**2
````

````python
myNums = [1,2,3,4,5]
````

````python
ans = list(map(square,myNums))
print(ans)
````

````python
list(map(lambda num: num ** 2, myNums))
````

###filter Function

````python
def checkEven(num):
    return num % 2 == 0
````

**Reshape**

````python
a = np.array([1,2,3,4,5,6,7,8])
print ('The original array:', a, '\n')

b = a.reshape(4,2)
print ('The modified array: \n', b)
````

**Linear Algebra**

````python
a = np.array([1,2,3,4,5,6]).reshape(2,3) 
print("Before Transpose: \n",a)

b = np.transpose(a)
print("\nAfter Transpose: \n",b)
````

````python
a = np.array([1,2,3,4]).reshape(2,2)
b = np.array([5,6,7,8]).reshape(2,2)
 
print("a: \n",a,"\nb: \n",b)
 
print("\nDot Product: \n",(np.dot(a,b)))
 
print("\nInner Product: \n",(np.inner(a,b)))
 
print("\nCross Product: \n",(np.matmul(a,b)))
 
print("\nDeterminant of array a is : ",(np.linalg.det(a)),"and of array b is : ",np.linalg.det(b))
 
print("\nInverse of array a is : \n",(np.linalg.inv(a)),"\n\nand of array b is : \n",np.linalg.inv(b))
````

**Slicing**

````python
a = np.array([1,2,3,4,5])

print(a,"\n")

print(a[:2],"\n")

print(a[2:],"\n")

b = np.array([[1,2,3],[3,4,5]])

print(b)

print("\n",b[1:,2:])

print("\n",b[:1,:2])
````
